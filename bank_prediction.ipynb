### 3: Dummy Variable

```python
cat_col=["job","marital","education","default","housing","loan","contact","month","poutcome"]
```

```python
#getting dummies variable for all required column also it will drop the first variable which is of no use
dummies = pd.get_dummies(bank[cat_col],drop_first=True)
dummies.info()
```

```python
#Concating dummie variable than has been created to bank datasetb
bank=pd.concat([bank,dummies],axis=1)
bank.head()
```

```python
bank.columns
```

Here we have to remove the column whose dummies has already been created

```python
#Dropping the column whose dummies has already been created
bank.drop(cat_col,axis=1,inplace=True)
```

```python
bank.shape
```

# Step 4: Splitting the Data into Training and Testing Sets

As you know, the first basic step for regression is performing a train-test split.

```python
# Putting feature variable to X
X = bank.drop("y", axis=1)

X.head()
```

```python
y=bank['y']
y
```

```python
#importing librbay
from sklearn.model_selection import train_test_split
np.random.seed(0)
X_train,X_test,y_train,y_test= train_test_split(X,y,train_size=0.7,test_size=0.3,random_state=100)
```

```python
X_train.head()
```

```python
X_train.columns
```

## ->Scaling the variable by using standarisation method 

```python
#importing StandarScaler
from sklearn.preprocessing import StandardScaler
#using scaling 
scaler=StandardScaler()
num_var=["age","balance","day","duration","campaign","pdays","previous"]
X_train[num_var]=scaler.fit_transform(X_train[num_var])
X_train
```
